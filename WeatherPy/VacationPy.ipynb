### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
location = cities_df[["Lat", "Lng"]]
location.head()
```

```python
humidity = cities_df["Humidity"]
humidity.head
gmaps.configure(api_key=g_key)
maps = gmaps.figure()
heatlayer = gmaps.heatmap_layer(location, weights=humidity, dissipating=False, max_intensity=500, point_radius=3)
maps.add_layer(heatlayer)
maps
```
